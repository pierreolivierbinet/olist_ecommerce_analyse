```python
import pandas as pd

clients = pd.read_csv('olist_customers_dataset.csv')

clients.head()
```

```python
clients.info()
```

```python
commandes = pd.read_csv('olist_orders_dataset.csv')

details = pd.read_csv('olist_order_items_dataset.csv')

commandes.head()
```

```python
df_final = pd.merge(clients, commandes, on='customer_id')

df_final.head()
```

```python
df_final.shape
```

```python
df_complet = pd.merge(df_final, details, on='order_id')

print(df_complet.shape)

df_complet.columns
```

```python
!pip install duckdb

import duckdb
```

```python
requete = """
SELECT
    customer_city,
    SUM(price) AS chiffre_affaires
FROM df_complet
GROUP BY customer_city
ORDER BY chiffre_affaires DESC
LIMIT 10
"""

resultat_sql = duckdb.query(requete).df()

resultat_sql
```

```python
df_complet['order_purchase_timestamp'] = pd.to_datetime(df_complet['order_purchase_timestamp'])

print(df_complet['order_purchase_timestamp'].dtype)

df_complet['mois_annee'] = df_complet['order_purchase_timestamp'].dt.to_period('M')

df_complet.head()
```

```python
df_complet.to_csv('olist_data_final.csv', sep=';', index=False)
```

```python
avis = pd.read_csv('olist_order_reviews_dataset.csv')

avis = avis[['order_id', 'review_score']]

df_complet = pd.merge(df_complet, avis, on='order_id')

print(df_complet.columns)
```

```python
requete_notes = """
SELECT
    customer_city,
    AVG(review_score) AS note_moyenne,
    COUNT(order_id) AS nb_commandes
FROM df_complet
GROUP BY customer_city
HAVING nb_commandes > 100
ORDER BY note_moyenne ASC
LIMIT 10
"""

notes_villes = duckdb.query(requete_notes).df()
notes_villes
```
